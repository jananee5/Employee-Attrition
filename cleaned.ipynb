```python
#feature engineering
# Define tenure categories based on 'YearsAtCompany'
def categorize_tenure(years):
    if years < 3:
        return 'Short-Term'
    elif 3 <= years <= 7:
        return 'Mid-Term'
    else:
        return 'Long-Term'

# Apply the function to create the tenure category
data['tenurecategory'] = data['yearsatcompany'].apply(categorize_tenure)

# Preview the data
print(data[['yearsatcompany', 'tenurecategory']].head(20))
```

```python
# Compute a weighted average for performance score
data['performancescore'] = (
    0.4 * data['jobsatisfaction'] +  # Weight for Job Satisfaction
    0.3 * data['performancerating'] +  # Weight for Performance Rating
    0.3 * data['jobinvolvement']  # Weight for Job Involvement
)

# Preview the data
print(data[['jobsatisfaction', 'performancerating', 'jobinvolvement', 'performancescore']].head())
```

```python
# Workload indicator based on overtime and weekly working hours
# Assuming 0 = no overtime, 1 = overtime, and 40 hours as standard work hours
data['workloadindicator'] = data['overtime'] * 1  # Binary workload indicator
# OR use scaled overtime hours if available:
# data['WorkloadIndicator'] = data['OvertimeHours'] / 40

# Preview the data
print(data[['overtime', 'workloadindicator']].head())
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Drop rows with NaN in X or y
X_clean = X.dropna()
y_clean = y.loc[X_clean.index]  # Make sure y matches X's index

# Apply SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_clean, y_clean)

print('Resampled dataset shape:', Counter(y_res))
```

```python
print(X_res.shape)
print(y_res.shape)
```

```python
#Splitting the data into Train and Test
from sklearn.model_selection import train_test_split
# Correctly split the resampled dataset
xtrain, xtest, ytrain, ytest = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier (random_state=42)
rf_model = rf.fit(xtrain, ytrain)
```

```python
importances = rf.feature_importances_

importances
```

```python
pd.DataFrame(importances).sort_values(ascending=False,by=0).index[0:5]
```

```python
X_res.columns[pd.DataFrame(importances).sort_values(ascending=False,by=0).index[0:5]]
```

```python
import pandas as pd

std = np.std([tree.feature_importances_ for tree in rf.estimators_], axis=0)
forest_importances = pd.Series(importances, index=X_res.columns)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Print classification metrics
print('\nClassification Report')
print(classification_report(ytest, y_predicted))
print("Accuracy:", accuracy_score(ytest, y_predicted))
print("F1 Score:", f1_score(ytest, y_predicted, average='weighted'))  # or 'macro' or 'micro' based on your need

# Generate confusion matrix
cm = confusion_matrix(ytest, y_predicted)

# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=True, yticklabels=True)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Calculate probabilities for the positive class
y_prob = rf_model.predict_proba(xtest)[:, 1]

# Compute AUC-ROC
auc = roc_auc_score(ytest, y_prob)
print(f"AUC-ROC Score: {auc:.2f}")

# Generate ROC curve data
fpr, tpr, thresholds = roc_curve(ytest, y_prob)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random Guess')
plt.title('ROC Curve', fontsize=16)
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
